```python
import pandas as pd
import streamlit as st
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px 
```

```python
path = 'https://frenzy86.s3.eu-west-2.amazonaws.com/python/data/immobili.csv'
```

```python
df = pd.read_csv(path)
df
```

```python
df.rename(columns={"medv": "price"}, inplace=True)
df
```

```python
df = df.drop(df[df.index == 506].index)
df
```

```python
df = df.astype(float)
```

```python
df.info()
```

```python
df
```

```python
X = df.drop(columns='price')
y = df['price']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, 
                                                    test_size = 0.2, 
                                                    random_state = 80
                                                    )
```

```python
X_train.shape, X_test.shape
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error

r2score = r2_score(y_test, y_pred)
mae = mean_absolute_error(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
rmse = mean_squared_error(y_test, y_pred, squared=False)

print('R2_score: ', r2score)
print('MAE: ', mae)
print('MSE: ', mse)
print('RMSE: ', rmse)
```

```python
model.coef_
```

```python
model.intercept_
```

```python
y_pred = model.predict(X_test)
```

```python
res_df = pd.DataFrame(data=list(zip(y_pred, y_test)),columns=['predicted', 'real'])
res_df
```

```python
res_df['error'] = res_df['real'] - res_df['predicted']
res_df
```
